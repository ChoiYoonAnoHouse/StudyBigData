### t-검정 / 회귀분석

### 라이브러리 import

```python
from scipy import stats
```

```python
# statsmodels  패키치 추가
!pip install statsmodels
```

##### 회귀분석을 위한 statsmodels 패키치 import

```python
from statsmodels.formula.api import ols, glm
```

```python
red_wine_quality = wine.loc[wine.type == 'red', 'quality']
red_wine_quality
```

```python
white_wine_quality = wine.loc[wine['type'] == 'white', 'quality']
white_wine_quality
```

```python
# t-검정 (두 그룹 사이의 차이)
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)
```

```python
# 선형회귀 분석
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
```

```python
regre_result = ols(Rformula, data=wine).fit()
regre_result.summary()
```

```python
regre_result = glm(Rformula, data=wine).fit()
regre_result.summary()
```

### OLS모델로 계산된 회귀분석 결과로 새 샘플 와인에 대한 품질 등급 예측

```python
# quality, type은 제거하고 컬럼 알파벳 순으로 정렬
sample1 = wine[wine.columns.difference(['quality', 'type'])]
sample1 = sample1[0:5][:]
sample1
```

```python
sample1_predict = regre_result.predict(sample1)
sample1_predict
```

```python
wine[0:5]['quality']
```

```python
# 새로운 와인에서 추출 된 데이터를 가지고 등급을 예측
#alcohol	chlorides	citric_acid	density	fixed_acidity	free_sulfur_dioxide	pH	residual_sugar	sulphates	total_sulfur_dioxide	volatile_acidity
data = {'alcohol':[9.0, 0.88], 'chlorides':[0.055, 0.04], 'citric_acid':[0.3,0.4], 'density':[0.996,0.91], 'fixed_acidity':[8.5, 8.1], 'free_sulfur_dioxide':[30.0,31.0], 'pH':[3.25,3.01], 'residual_sugar':[6.1,5.8], 'sulphates':[0.4,0.35], 'total_sulfur_dioxide':[98.,99], 'volatile_acidity':[0.8,0.5]}

new_sample = pd.DataFrame(data, columns=sample1.columns)
new_sample
```

```python
new_sample_predict = regre_result.predict(new_sample)
```

```python
new_sample_predict
```

### 시각화

#### 와인 유형에 따른 품질 등급 히스토그램

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('dark')
plt.rcParams['font.family'] = 'MaruBuri'
sns.distplot(red_wine_quality, kde=True, color='red', label='레드 와인')
sns.distplot(white_wine_quality, kde=True, color='blue', label='화이트 와인')
plt.title('와인 타입 품질 등급')
plt.legend()

plt.show()
```

#### 선형회귀 분석 값 시각화
