##### 변수 사이 상관계수를 히트맵으로 시각화

```python
def categroy_age(x): # 연령대 가져오기 함수
    if x < 10: return 0
    elif x < 20: return 1
    elif x < 30: return 2
    elif x < 40: return 3
    elif x < 50: return 4
    elif x < 60: return 5
    elif x < 70: return 6
    else: return 7
```

```python
titanic
```

```python
titanic['age_band'] = titanic['age'].apply(categroy_age)
```

```python
titanic['sex'] = titanic['sex'].map({'male': 1, 'female': 0})
```

```python
titanic['family'] = titanic['sibsp'] + titanic['parch'] + 1 # +!의 이유는 혼자인 사람이 있기 때문 ! sibsp, parch는 형제 이기 때문에 혼자인 사람은 0으로 떠서 + 1 하는 것.
```

```python
titanic.to_csv('./data/titanic_datahandling.csv', index=False, encoding='utf-8')
```

```python
# 히트맵 그리기
heatmap_data = titanic[['survived', 'sex', 'age_band', 'family', 'pclass', 'fare']]
color_map = plt.cm.RdBu
#음수 마이너스가 깨질 때
plt.rcParams['axes.unicode_minus'] = False
sns.heatmap(data=heatmap_data.astype(float).corr(), linewidths=0.1, square=True, cmap=color_map, annot=True)
 #corr 상관계수를 분석하는 함수, square 사각형, cmap은 색을 바꿀 때 사용, annot는 사각형 안에 숫자가 나옴.
 #높을 수록 색이 진하다. 음수가 제대로 안 나올 경우 처리해줘야 한다.
plt.show()
```
