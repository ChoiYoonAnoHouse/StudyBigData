```python
# 파일 병합
all_files = glob.glob('./myCabinetExcelData*.xls') # *을 사용하면 모든 파일을 추출
all_files 
```

```python
all_data = []
for file in all_files:
    df = pd.read_excel(file)
    all_data.append(df)

all_data[0]
```

```python
all_data_concat = pd.concat(all_data, axis=0, ignore_index=True)
all_data_concat
```

```python
all_data_concat = all_data_concat.drop(all_data_concat.columns[[0]], axis='columns')
all_data_concat
```

```python
all_data_concat.to_csv('./riss_bigdata.csv', index=False, encoding='utf-8')
```

#### 데이터 전처리

```python
all_title = all_data_concat['제목']
all_title
```

```python
stopWords = set(stopwords.words('english'))
stopWords
```

```python
lemma = WordNetLemmatizer()
```

```python
import re

words = []
 # +는 이 정규식이 계속 된다는 뜻이다.
for title in all_title:
    EnWords = re.sub(r'[^a-zA-Z]+', ' ', str(title)) # '' 사이에 공백이 꼭 들어가야한다.
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop] #표제어 추출

    words.append(EnWordsTokenStopLemma)
```

```python
print(words) # 불용어 제거 남은 키워드 단어만 남음
```

```python
# 2차원 리스트 -> 1차원 리스트로
words_reduce = list(reduce(lambda x,y : x+y, words))
print(words_reduce)
```

```python
#Counter은 collections안에 있는 함수
count = Counter(words_reduce)
count
```

```python
word_count = dict() #dict는딕셔너리

for tag, counts in count.most_common(30):
    if len(str(tag))> 1:
        word_count[tag] = counts
        # print(f'{tag}: {counts}')

word_count
```

#### 단어 개수 차트

```python
sorted_Keys = sorted(word_count, key=word_count.get, reverse=True) # reverse 큰거에서 작은수로
sorted_Values = sorted(word_count.values(), reverse=True)

plt.figure(figsize=(20, 5))
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation=90)
plt.show()
```

#### 결과 시각화

##### 워드 클라우드

```python
stopwords = set(STOPWORDS)
wc = WordCloud(background_color='#F5F5F5', stopwords=stopwords, width=1280, height=768)
cloud = wc.generate_from_frequencies(word_count)

plt.figure(figsize=(40, 15))
plt.imshow(cloud)
plt.axis('off') #좌표
plt.show()
```
